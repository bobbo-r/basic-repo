```python
#find parabola, error, line of best fit for a scatterplot?
#error: 1/n * sum((yobs-yexp)^2)
#From OpenAI:

import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.random.rand(50) * 10
y = 2 * x + 1 + np.random.randn(50) * 2  # y = 2x + 1 + noise

# Calculate the coefficients of the line of best fit (linear regression)
coefficients = np.polyfit(x, y, 1)
m, b = coefficients  # slope (m) and y-intercept (b)

# Calculate the predicted y values using the fitted line
y_pred = m * x + b

# Calculate the error (root mean squared error)
error = np.sqrt(np.mean((y - y_pred)**2))

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Data')

# Plot the line of best fit
plt.plot(x, y_pred, color='red', label=f'Line of Best Fit: y = {m:.2f}x + {b:.2f}')

# Calculate and display the error
plt.text(0.5, 25, f'RMSE: {error:.2f}', fontsize=12)

# Customize plot
plt.title('Scatter Plot with Line of Best Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
print(error)
```

```python
# sqrt(MSE) = RMSE root mean square

# MSE = 1/n * sum((return val-actual val of function)^2) mean square error

#chi-sqwaured chi^2 = sum (observed data point y-f(xi)/sigma.i) (error at point i)

#packages to use:
"""
numpy.ployfit(x_array, y_array, degree)
scipy.optimize.curve_fit(eq, x, y, p0 (initial guess ))
"""

```

```python
#DEMO

import matplotlib.pyplot as plt
import scipy.optimize as fit
m=.5
b=3
sigma=2
n=20 #20 points

xdata = np.arange(0, n)
ydata = m*xdata + b + sigma*np.random.standard_normal() #makes y value differing from line of best fit by sigma*randomnormal fuinction (noise)
ydata
```

```python
plt.figure(figsize=(8,6))
plt.scatter(xdata, ydata)
plt.show()
```

```python
def f(x, m, b): # must have x first
    return m*x+b
```

```python
p0=[0.3, 1]
parameters, covariance_matrix = fit.curve_fit(f, xdata, ydata, p0)
```

```python
covariance_matrix # shows errors for each point, params are slope and intercept
```

```python

m_error = covariance_matrix[0,0]**.5
b_error = covariance_matrix[1,1]**.5
m=parameters[0]
b=parameters[1]
print('m={0:1.5f} +/- {1:1.5f}'.format(m, m_error))
```

```python
x=np.linspace(xdata.min(), xdata.max(), 100)
y=f(x, m, b)
plt.figure()
plt.plot(x, y)
plt.scatter(xdata, ydata)
plt.show() #plots regression line
```
